```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
def plot_graph(x,y,title="Data",scatter=True):
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")

    if scatter:
        plt.scatter(x,y)
    else:
        plt.plot(x,y)
    
    plt.axis('equal')
    plt.show()
```

```python
plot_graph(x,y)
```

```python
def draw_line(m,c):
    x = [0,1,2,3,4,5]
    y = []

    for item in x:
        y.append(((m*item)+c))

    plot_graph(x,y,scatter=False)

    return y
```

```python
draw_line(23,100)
```

```python
def cost_function(y_hat_arr,y_arr):
    n = len(y_arr)

    deviation = 0 

    for y_hat,y in zip(y_hat_arr,y_arr):
        deviation += (y_hat - y)**2

    return deviation/n
```

```python
# Let's take a scenario 
# where c=0 


# Consider a best fit line passing through origin 
```

```python
# Consider three cases  

# Let m=0.5,1,2

y_hat_arr = [0,1,2,3,4,5]  # this is the best fit line
m = [0.5,1,2]
cost_function_arr = []

y_case1 = draw_line(m[0],c=0)
cost_function1 = cost_function(y_hat_arr,y_case1)
cost_function_arr.append(cost_function1)


y_case2 = draw_line(m[1],c=0)
cost_function2 = cost_function(y_hat_arr,y_case2)
cost_function_arr.append(cost_function2)


y_case3 = draw_line(m[2],c=0)
cost_function3 = cost_function(y_hat_arr,y_case3)
cost_function_arr.append(cost_function3)
```

```python
plot_graph(cost_function_arr,m)
```

```python
# You can Observe the Gradient Descent Curve 
# In This Case, we already knew the Best Fit line Slope( ie: m=1)
# In general case, where we have to find the best fit line 
# We take a random point in the space 
# And apply Convergence theorem untill we get Good result


# Convergence Theorem

# m = m - learning_rate * partial_differentiation_of_E_wrt_m
# b = b - learning_rate * partial_differentiation_of_E_wrt_b
```

```python
def gradient_descent(m_now,b_now,n,learning_rate,x,y):
    m_gradient,b_gradient = 0,0 

    for i in range(n):
        m_gradient += -(2/n) * x[i] * (y[i] - ((m_now * x[i]) + b_now))
        b_gradient += -(2/n) * (y[i] - ((m_now * x[i]) + b_now))
        
    m = m_now - m_gradient * learning_rate
    b = b_now - b_gradient * learning_rate


    return m,b
```

```python
def draw_best_fit_line(X,Y,m_new,b_new):
    plt.title("Best Fit Line")
    plt.xlabel("X")
    plt.ylabel("Y")

    plt.scatter(X,Y)

    x_new = [0,1,2,3,4,5]
    y_new = []

    for item in x_new:
        y_new.append(((m_new*item)+b_new))

    plt.plot(x_new,y_new)
    plt.show()
    
```

```python
x = [5.48813504, 7.15189366, 6.02763376, 5.44883183, 4.23654799]
y = [12.32765641, 16.08010966, 14.09100814, 11.48492694,  8.65574123]


epochs = 1000
m = 0 
b = 0 
L = 0.001

for i in range(epochs):
    m, b = gradient_descent(
        m_now=m,
        b_now=b,
        n=5,
        learning_rate=L,
        x=x,
        y=y
    )

print(f"m={m}\nb={b}")
draw_best_fit_line(x,y,m,b)
```
